# How to organize data in Python

Remember I mentioned that in matlab you can organize the data in numerical matrix, cell arrays, structure arrays or tables. 
Python works differently and has different ways of organizing the data. In Python, you can organize data using different data structures:

```python
# 1. Lists: Ordered, mutable collections
numbers = [1, 2, 3, 4, 5]

# 2. Tuples: Ordered, immutable collections
coordinates = (10.0, 20.0)

# 3. Dictionaries: Key-value pairs, unordered
person = {"name": "Alice", "age": 30}

# 4. Sets: Unordered collections of unique elements
unique_numbers = {1, 2, 3}

print("List:", numbers)
print("Tuple:", coordinates)
print("Dictionary:", person)
print("Set:", unique_numbers)
```

#### Why use a list?

Lists are useful because they are ordered, mutable (can be changed), and can store elements of different types.
You can easily add, remove, or modify elements in a list.
Example: adding a new number to the 'numbers' list

```python
numbers.append(6)
print("Updated list:", numbers)
```

#### Why use tuple instead of list?
Once created, a tuple cannot be changed (no adding, removing, or changing elements).
Hashable: Tuples can be used as dictionary keys or elements of sets, while lists cannot.
| Feature             | Tuple                            | List                     |
| ------------------- | -------------------------------- | ------------------------ |
| Mutable             | ❌ No                             | ✅ Yes                    |
| Hashable            | ✅ Yes (if contents are hashable) | ❌ No                     |
| Faster              | ✅ Yes (slightly)                 | ❌ Slower                 |
| Use as dict key/set | ✅ Yes                            | ❌ No                     |
| Typical use case    | Fixed, structured data           | Dynamic, changeable data |


#### When to use a dictionary?
You should use a dictionary in Python when you want to store and retrieve values by a key, rather than by index like with lists or tuples. Dictionaries are ideal for mapping and fast lookups.

# How to use conditional statement and for loops

### 1. Conditional statement
A conditional statement is a way to make decisions in your code.
It allows you to execute certain code only if a specific condition is true.

#### **Example** Check if a number is even or odd


```python
num = 2
if num % 2 == 0:  #the expression num % 2 means you're taking the remainder when num is divided by 2. This is called the modulo operator (%).
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")
```

### 2. For loop
A **for loop** in Python is a control structure that allows you to repeat a block of code for each item in a sequence (such as a list, tuple, string, or range). It is commonly used to iterate over collections of data and perform actions on each element.


#### **Example** Run the same code for a list of numbers

```python
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    if num % 2 == 0:
        print(f"{num} is even.")
    else:
        print(f"{num} is odd.")
```

## **Example 3**
You will generate two distributions and plot them using different methods
1. you will generate 2 random distributions with 100 values using numpy. data 1 is centered at 0 with std of 1, data 2 is centered at 2 with std of 1
2. you will plot them as a histogram
3. you will plot them as a boxplot
4. you will plot them as a violin plot
5. you will save the data as an excel sheet and a csv file

```python
# Generate two lists with 100 random values each, with significantly different means
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=5, scale=1, size=100)

# Plot histogram
plt.hist(data1, bins=20, alpha=0.5, label='data1')
plt.hist(data2, bins=20, alpha=0.5, label='data2')
plt.title("Histogram of Two Different Distributions")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Plot box plot
plt.boxplot([data1, data2], tick_labels=['data1', 'data2'])
plt.title("BoxPlot of Two Different Distributions")
plt.ylabel("Value")
plt.show()

# Plot violin plot
plt.violinplot([data1, data2], showmeans=True)
plt.xticks([1, 2], ['data1', 'data2'])
plt.title("Violin Plot of Two Different Distributions")
plt.ylabel("Value")
plt.show()
```

```python
# make sure to install the required libraries
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# The ANOVA test (f_stat, p_val) tells you if there is a statistically significant difference between the means of the groups (exp1, exp2, exp3).
# If the p-value is less than your significance level (commonly 0.05), you can conclude that at least one group mean is different.

# Perform one-way ANOVA test
f_stat, p_val = f_oneway(df['exp1'], df['exp2'], df['exp3'])
print(f"ANOVA F-statistic: {f_stat:.3f}, p-value: {p_val:.3e}")

# However, ANOVA does not tell you which groups are different from each other.

# To find out which specific groups differ, you need to perform a post-hoc multiple comparison test.
# A common choice is Tukey's HSD (Honestly Significant Difference) test.


# Combine all data into a single array and create a group label array
data = np.concatenate([df['exp1'], df['exp2'], df['exp3']])
groups = (['exp1'] * len(df['exp1'])) + (['exp2'] * len(df['exp2'])) + (['exp3'] * len(df['exp3']))

# Perform Tukey's HSD test
tukey_result = pairwise_tukeyhsd(data, groups)
print(tukey_result)
```

```python
def scatter_on_violin(data, ax=None, color='blue', scatter_kwargs=None, violin_kwargs=None, label=None):
    """
    Plots a violin plot with a scatter overlay for a list or numpy array of numbers.

    Parameters:
        data (list or np.ndarray): The data to plot.
        ax (matplotlib.axes.Axes, optional): The axes to plot on. If None, uses current axes.
        color (str, optional): Color for scatter points.
        scatter_kwargs (dict, optional): Additional kwargs for plt.scatter.
        violin_kwargs (dict, optional): Additional kwargs for plt.violinplot.
        label (str, optional): Label for scatter points.
    """
    if ax is None:
        ax = plt.gca()
    if scatter_kwargs is None:
        scatter_kwargs = {}
    if violin_kwargs is None:
        violin_kwargs = {}

    # Violin plot
    parts = ax.violinplot([data], showmeans=True, **violin_kwargs)
    for pc in parts['bodies']:
        pc.set_facecolor('lightgray')
        pc.set_alpha(0.5)

    # Scatter overlay with jitter
    x_jitter = np.random.normal(1, 0.04, size=len(data))
    ax.scatter(x_jitter, data, color=color, alpha=0.7, label=label, **scatter_kwargs)
    ax.set_xticks([1])
    ax.set_xticklabels([label if label else "Data"])
    ax.set_title("Scatter on Violin Plot")
    ax.set_ylabel("Value")
    if label:
        ax.legend()

# Example usage:
# scatter_on_violin(data1, label='data1')
# plt.show()
# Example usage of the scatter_on_violin function


scatter_on_violin(df['exp2'])
```
